```python
#import pandas
import pandas as pd
```

```python
#read in data for analysis
white_wine_data = pd.read_csv('winequality-white.csv',sep=';')
white_wine_data.tail()
```

```python
bins = (2, 6.5, 8)
group_names = ['bad', 'good']
white_wine_data['quality'] = pd.cut(white_wine_data['quality'], bins = bins, labels = group_names)
```

```python
white_wine_data['quality'] = pd.get_dummies(white_wine_data['quality'])
white_wine_data.head()
```

```python
#specify the data to be used in running the model
X = white_wine_data[['chlorides','alcohol','sulphates','volatile acidity','citric acid']]
y = white_wine_data['quality'].reshape(-1,1)
print(X.shape)
print(y.shape)
print(white_wine_data['quality'].unique())
```

```python
#split data for training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
#scale data and normalize
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
from keras.utils import to_categorical

# One-hot encoding
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
y_test_categorical[0]
```

```python
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(units=10,activation='relu',input_dim=5))
model.add(Dense(units=10,activation='relu'))
model.add(Dense(units=10,activation='relu'))
model.add(Dense(units=7,activation='relu'))
model.add(Dense(units=7,activation='relu'))
model.add(Dense(units=5,activation='relu'))
model.add(Dense(units=2,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', 
              loss='binary_crossentropy', 
              metrics=['accuracy'])
```

```python
model.fit(X_train_scaled,y_train_categorical,epochs=100,shuffle=True,verbose=1)
```

```python
pred = model.predict(X_test)
pred
```

```python
model_loss, model_accuracy = model.evaluate(X_test_scaled, y_test_categorical, verbose=2)
print(f"Deep Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```
